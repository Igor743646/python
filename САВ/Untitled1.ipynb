```python
def f(x):
    if x == 1:
        return 1
    elif x % 2 == 0 and x % 3 == 0:
        return f(x-1) + f(x//2) + f(x//3)
    elif x % 2 == 0:
        return f(x-1) + f(x//2)
    elif x % 3 == 0:
        return f(x-1) + f(x//3)
    else:
        return f(x-1)


a = point([(i, f(i)) for i in range(1, 10)])
```

```python
b = point([(i, 3**(i)) for i in range(1, 10)], color='red')
plot(a) + plot(b)
```
