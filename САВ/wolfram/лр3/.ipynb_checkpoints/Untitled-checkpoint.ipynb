<h2>Модель "Хищник - жертва"</h2>

```
x = .;
y = .;
grafic[aa_, cc_, dd_] := Module [{a = aa, c = cc, d = dd, plt = {}},
    For[b = 4, b > 1, b--,
        eq1[t_] := x'[t] == (a - b * y[t]) * x[t];
        eq2[t_] := y'[t] == (- c + d * x[t]) * y[t];
        solution = NDSolve[{eq1[t], eq2[t], x[0] == 2, y[0] == 1}, {x, y}, {t, 0, 7}];
        plt = Append[plt, ParametricPlot[Evaluate[{x[t], y[t]} /. solution], {t, 0, 7}, PlotRange -> {{0, 3}, {0, 3}}]];
    ];
    plt
];
```

```
ListAnimate[Table[Show[grafic[3, n, 1]], {n, 0.3, 1, 0.05}]]
(*Manipulate[Show[grafic[3, n, 1]], {n, 0.5, 1, 0.01}] (*работает в wolfram cloud*)*)
```
