<h2>Упростить уравнение поверхности второго порядка в простраснтве</h2>

```
f[x_, y_, z_] := -2y^2 + 4y*z - 3z^2 + 4y + 4z - 12
f[x,y,z] // TraditionalForm
ContourPlot3D[f[x,y,z] == 0, {x, -4, 4}, {y, 0, 10}, {z, 0, 7}, 
                ContourStyle -> Directive[Orange, 
                                          Opacity[0.5], 
                                          Specularity[White, 60]
                                         ]
             ]
```

```
A := {{0, 0, 0}, {0, -2, 2}, {0, 2, -3}}
MatrixForm[A]
MatrixForm[Transpose[Eigenvectors[A]]]
```

```
a := {0, 2, 2}
MatrixForm[a]
```

```
a0 = -12 
```

```
cp = Det[A - l*IdentityMatrix[3]]
CharacteristicPolynomial[A, l] == cp
```

```
rcp := l /. Solve[cp == 0, l]
rcp3 := rcp[[1]]
rcp2 := rcp[[2]]
rcp1 := rcp[[3]]
MatrixForm[{N[rcp1], N[rcp2], N[rcp3]}]
(* Sort[Eigenvalues[A], Greater] === Sort[{rcp1, rcp2, rcp3}, Greater] *)
```

```
l1 := Solve[(A - rcp1*IdentityMatrix[3]).{a1, a2, 4} == {0,0,0}]
l2 := Solve[(A - rcp2*IdentityMatrix[3]).{a1, a2, 4} == {0,0,0}]
l3 := Solve[(A - rcp3*IdentityMatrix[3]).{1, a2, a3} == {0,0,0}]
l1 = {a1,a2,a3} /. Flatten[Append[{a3 -> 4}, l1[[1]]]]
l2 = {a1,a2,a3} /. Append[l2[[1]], a3 -> 4]
l3 = {a1,a2,a3} /. Append[l3[[1]], a1 -> 1]

Eigensystem[A] == {Sort[{rcp1, rcp2, rcp3}, Less], {l2,l3,l1}}
```

```
s1 = FullSimplify[Normalize[l1]]
s2 = FullSimplify[Normalize[l2]]
s3 = FullSimplify[Normalize[l3]]
```

```
S := Transpose[{s1,s2,s3}]
MatrixForm[N[S]]
```

```
al = N[FullSimplify[Transpose[S].a]]
```

```
ff[xx_, yy_, zz_] := rcp1*xx^2 + rcp2*yy^2 + rcp3*zz^2 + 
                                    2*al[[1]]*xx + 2*al[[2]]*yy + 2*al[[3]]*zz + a0
ff[xx, yy, zz] // TraditionalForm

ContourPlot3D[ff[xx+7, yy, zz] == 0, {xx, -9, 9}, {yy, -9, 9}, {zz, -7, 7}, 
                ContourStyle -> Directive[Orange, 
                                          Opacity[0.5], 
                                          Specularity[White, 60]
                                         ]
             ]
```
